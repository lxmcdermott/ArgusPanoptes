# Argus Panoptes — Synthetic Signals Exploration (placeholder)

Quick sandbox for poking at the `sensors/` simulators. Run from repo root.


```python
import numpy as np
import matplotlib.pyplot as plt

from sensors import SawVibrationSimulator, ThermalSimulator, load_config

cfg = load_config()
vib = SawVibrationSimulator(cfg)
therm = ThermalSimulator(cfg)

t, accel, meta = vib.generate(duration_s=1.0, wear=0.2, seed=0)
print('TPF (Hz):', meta['tooth_pass_freq_hz'])
plt.figure(figsize=(10, 3))
plt.plot(t[:4000], accel[:4000])
plt.xlabel('time (s)'); plt.ylabel('accel (g)'); plt.title('Vibration @ wear=0.2')
plt.tight_layout(); plt.show()
```
